```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science'])
plt.rcParams.update({'font.size': 20, 'font.family': 'sans-serif'})
plt.rcParams.update({
        "font.weight": "bold",  # bold fonts
        # "xtick.labelsize": 15,   # large tick labels
        # "ytick.labelsize": 15,   # large tick labels
        "lines.linewidth": 1,   # thick lines
        "lines.color": "k",     # black lines
        "grid.color": "0.5",    # gray gridlines
        "grid.linestyle": "-",  # solid gridlines
        "grid.linewidth": 0.5,  # thin gridlines
        "savefig.dpi": 300,     # higher resolution output.
    })
```

```python
def LJP(r,sigma,u):
    """Lennard-Jones Potential"""
    return u*((sigma/r)**12-(sigma/r)**6)
def hard_sphere(r,sigma,u):
    """Hard Sphere Potential"""
    return -u/r**6

r = np.linspace(0.9,3,100)
sigma = 1.0
u = 1.0

r1 = np.linspace(2**(1/6)*sigma,3,100)
```

```python
fig,axes = plt.subplots(1,2, figsize=(20,10))
ax= axes[0]
ax.plot(r, LJP(r,sigma,u), label=r"$\sigma$" +f"{sigma},"+r"$ u=$"+f"{u}", color='teal', lw = 2)
ax.axhline(0, color='black', lw=1, ls='-')
ax.axvline(sigma*2**(1/6), color='red', lw=0.5, ls='-', label=r'$r_{min}=\sigma 2^{1/6}$')
ax.set_title('Lennard-Jones Potential')
ax.annotate(r'$\sim r^{-12}$', xy=(1, 0.8), xytext=(1., 0.8), fontsize=25)
ax.annotate(r'$\sim r^{-6}$', xy=(2.7, -0.1), fontsize=25)
ax.set_xlabel('Distance r')
ax.set_ylabel('Potential Energy U(r)')
ax.set_ylim(-0.5,1)
ax.set_xlim(0.8,3)

ax= axes[1]
ax.plot(r1, hard_sphere(r1,sigma,u), label=r"$\sigma$" +f"{sigma},"+r"$ u=$"+f"{u}", color='teal', lw = 2)
ax.axhline(0, color='black', lw=1, ls='-')
ax.vlines(2**(1/6)*sigma, ymin=hard_sphere(r1[0],sigma,u), ymax=1, color='teal', lw=2, ls='-', label=r'$r_{min}=\sigma 2^{1/6}$')
ax.set_title('Hard Sphere Potential')
ax.annotate(r'$\sim r^{-6}$', xy=(2.7, -0.1), fontsize=25)
ax.set_xlabel('Distance r')
ax.set_ylabel('Potential Energy U(r)')
ax.set_ylim(-0.7,1)
ax.set_xlim(0.8,3)
# plt.legend()

# fig.savefig('images/LJ_and_HS_Potentials.png', dpi=300)
```
