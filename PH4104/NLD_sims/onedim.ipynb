```python
import numpy as np
import matplotlib.pyplot as plt 
import scienceplots
plt.style.use(['science'])
plt.rcParams.update({'figure.dpi': '150'})
plt.rcParams.update({'font.size': 15, 'font.family': 'serif'})
```

```python
X,Y = np.meshgrid(np.linspace(0,3,100),np.linspace(-3,3,200))
U = X
MU = -2
V = MU*Y - Y**3
# V = MU*Y - Y**2
```

```python
fig,axes = plt.subplots(figsize=(8,6))
axes.streamplot(X,Y,U,V,density=3,linewidth=1,arrowstyle='-|>', cmap = 'viridis')
# axes.plot(xvals2, yvals2, color='red', linewidth=2)
# axes.plot(xvals3, yvals3, color='red', linewidth=2)
# axes.scatter([0,3],[2,0],color='red', zorder=5)
# axes.scatter([0,1],[0,1],color='green', zorder=5)
axes.set_xlim(-0.02,3)
# axes.set_ylim(-0.02,5)
axes.grid()
axes.set_xlabel('t')
axes.set_title(r"$\mu = -2$")
axes.set_ylabel(r'$\frac{d x}{d t} = \mu x-x^3 $')
axes.axhline(y=-np.pow(MU,1/2), color='green', linestyle='--', linewidth=3)
axes.axhline(y=np.pow(MU,1/2), color='green', linestyle='--', linewidth=3)
axes.axhline(y=0, color='green', linestyle='--', linewidth=3)
fig.savefig('images/oned_phase_pitchfork_bef.png',bbox_inches='tight')
```

```python
fig,axes = plt.subplots(figsize=(8,8))
xvals1 = np.linspace(-5,0,1000)
xvals2 = np.linspace(0,5,1000)
yvals21 = np.sqrt(xvals2)
yvals22 = -np.sqrt(xvals2)
yvals23 = 0*xvals2
yvals1 = 0*xvals1
# yvals2 = np.sqrt(xvals2)
axes.plot(xvals1, yvals1, color='red', linewidth=2, label=r'Unstable fixed point')
axes.plot(xvals2, yvals23, color='red', linewidth=2, label=r'Unstable fixed point')
axes.plot(xvals2, yvals21, color='green', linewidth=2, label=r'Stable fixed point')
axes.plot(xvals2, yvals22, color='green', linewidth=2, label=r'Stable fixed point')
axes.set_xlim(-5,5)
axes.set_ylim(-5,5)
# axes.grid()
# axes.legend()
axes.set_xlabel(r'$\mu$')
axes.set_ylabel(r'$\overline{x}$')
fig.savefig('images/oned_bifurcation_pitchfork.png',bbox_inches='tight')
```
