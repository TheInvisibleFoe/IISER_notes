```python
import numpy as np
import matplotlib.pyplot as plt
# import scienceplots
```

```python
X,Y = np.meshgrid(np.linspace(0,5,100),np.linspace(0,5,100))
U = X*(3-X-2*Y)
V = Y*(2-X-Y)
```

```python
fig,axes = plt.subplots(figsize=(6,6))
axes.streamplot(X,Y,U,V,density=1.5,linewidth=1,arrowstyle='-|>',color='blue')
axes.scatter([0,3],[2,0],color='red')
axes.scatter([0,1],[0,1],color='green')
axes.set_xlabel('X')
axes.set_ylabel('Y')
```

```python
X1,Y1 = np.meshgrid(np.linspace(-5,5,1000),np.linspace(-5,5,1000))
MU = 2
def vposc(X,Y,m):
    dXdt = m*(X - (X**3)/3 - Y)
    dYdt = X/m
    return dXdt, dYdt

def vposc2(X,Y,m):
    dXdt = Y
    dYdt = m*(1 - X**2)*Y - X
    return dXdt, dYdt
```

```python
fig, axes = plt.subplots(figsize=(8,8))
# axes.streamplot(X1, Y1, *vposc(X1, Y1, 2), density=1.5, linewidth=1, arrowstyle='-', color='purple')
axes.streamplot(X1, Y1, *vposc(X1, Y1, 1), density=3.5,linewidth=0.5, arrowstyle='-', color='black')
axes.set_xlabel('X')
axes.set_ylabel('Y')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.streamplot(X1, Y1, *vposc2(X1, Y1, MU), density=3.5, linewidth=0.5, arrowstyle='-', color='black')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
def selkov(X,Y,a,b):
    dXdt =  - X +a*Y + (X**2)*Y
    dYdt = b - a*Y - (X**2)*Y
    return dXdt, dYdt
```

```python
X2,Y2 = np.meshgrid(np.linspace(0,3,1000),np.linspace(0,3,1000))
fig, ax = plt.subplots(figsize=(8,8))
ax.streamplot(X2, Y2, *selkov(X2, Y2, 0.08, 0.6), density=1.5, linewidth=0.5, arrowstyle='-', color='green')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```
