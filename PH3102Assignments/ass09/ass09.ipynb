```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
V0 = 1
a = 1
m = float(5)
x = np.linspace(0,a,1000)
h0x = np.sqrt(2/a)*np.sin(np.pi*x/a)
# let a/hbar = 1
h1x = h0x + 3*np.sqrt(3)*V0*m/(16 * np.pi**3) * np.sqrt(2/a)*np.sin(3*np.pi*x/a)
h2x = h0x - 3*np.sqrt(3)*V0*m/(16 * np.pi**3) * np.sqrt(2/a)*np.sin(3*np.pi*x/a)
fig,axes = plt.subplots(1,1,figsize=(12,9))
plt.plot(x,h0x, label=r"Unperturbed Potential",ls="--",lw=3)
plt.plot(x,h1x, label=r"$V_0 > 0$")
plt.plot(x,h2x, label=r"$V_0<0$")
plt.axvline(1/3, ls ="--", color = "k")
plt.axvline(2/3, ls ="--", color = "k")
plt.xlim([0,1])
plt.ylim(bottom=0)
plt.xticks([0,0.2,0.33333,0.5,0.6666666,0.8,1.0],["0",r"$a/5$",r"a/6",r"a/2",r"2a/3",r"4a/5",r"a"])
plt.legend()
plt.grid()
plt.ylabel(r"$\psi (x)$")
plt.xlabel("Position(x)")
fig.savefig("wavefunc.png", dpi =600)
```
